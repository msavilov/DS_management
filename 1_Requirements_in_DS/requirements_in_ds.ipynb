### Описание задания:
На основе информационного продукта "Профессия Data Scientist" от Нетология, рассмотреть основные метрики продукта и подобрать состав команды для его развития.


### Задания:

1) Предложить способ организации аналитики по продукту.
2) Сформулировать 2–3 ключевые метрики продукта.
3) Определить какие данные могут помочь вам в поиске решения.
4) Описать команду для работы с данными для этого продукта.
5) Определить какие сотрудники будут нужны.
6) Описать задачи сотрудников.

### Cпособ организации аналитики по продукту

Так как Нетология крупная компания и продукт "Профессия Data Scientist" - один из многих, то оптимальным был бы гибридный способ организации команды аналитики.
Аналитику по продукту можно условно разбить на две части: 

- маркетинговую составляющую, которая будет описывать эффективность маркетинговых мероприятий по привлечению новых клиентов и ситуации на рынке продукта в целом;
- клиентскую аналитику, которая будет описывать действия клиентов по взаимодействию с продуктом и удовлетворённость клиентов;

Рассмотрим ситуацию, когда в компании маркетинговая аналитика является ответственностью отдела маркетинга и рассмотрим только клиентскую аналитику.


Этапы работы:
- определение бизнес потребностей и метрик продукта;
- постановка задач команде от владельца продукта;
- получение необходимых данных командой;
- анализ данных аналитиками;
- принятие решений на основе данных от аналитиков;
- реализация решений разработчиками.

В процессе использования клиентом продукта проводится сбор данных о прохождении им курсов, количестве выполненных домашних заданий, удовлетворенности занятиями и учебными модулями, посещаемость вебинаров. 

Данные хранятся в централизованной системе хранения данных компании, специалист по БД обеспечивает загрузку собранных данных из опросов клиента после занятий и быструю выгрузку необходимых данных.

### Определение ключевых метрики продукта

CSI (индекс удовлетворенности) - удовлетворенность клиентов продуктом. Опросы клиентов по качеству материала, после занятия и учебного модуля. 

NPS (индекс потребительской лояльности) - общая удовлетворённость клиента брендом и готовность к повторным приобретениям .

COR (показатель завершенности) - доля клиентов, завершивших курсы.

Дополнительной метрикой следует рассматривать долю клиентов, которые успешно нашли работу по профессии после окончания курса. Она получается от команды трудоустройства и характеризует общую ценность продукта для рынка специалистов в этой области.

### Описание данных, которые могут помочь вам в поиске решения.

Время проверки домашних заданий и ответов на обращения в обратную связь, позволит отследить необходимость привлечения дополнительных сотрудников для работы над продуктом.

Количество дополнительных вопросов по заданиям, позволит определить ясность и полноту описания задания для клиентов.

Определить оптимальную сложность курсов помогут следующие данные:
- доля клиентов, успешно завершивших курсы и получивших диплом;
- отслеживание прогресса клиентов и время на выполнение заданий;
- доля выполненных заданий, по которым необходима доработка;
- время затрачиваемое для просмотра лекции/семинара;
- вовзращение клиента к уже пройденным им темам;

Показатели обратной связи студентов.

### Состав и задачи команды

- Владелец продукта: организация работы команды и формирования бизнес требований к продукту;

- Разработчики: разработки API, тестирование и проведение интеграций;

- Специалисты по БД: обеспечение быстрого и надёжного доступа к централизованной системе хранения данных компании;

- Аналитик данных: построение моделей, позволяющие предсказывать изменения метрик; формулирование гипотез, поиск решений
